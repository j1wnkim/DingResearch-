```python
def notNumericData(data):
    nonNumericColumns = []
    for i in range(len(data.columns)):
        if data.iloc[:,i].dtype != "float64" and data.iloc[:,i].dtype != "int64":
            nonNumericColumns.append(data.columns[i])
    return nonNumericColumns


nonNumericColumns = notNumericData(LSTMData)
nonNumericColumns = nonNumericColumns[1:]
print(nonNumericColumns)
```

```python
def splitData(data, output):
    dataCopy = data.copy()
    outputCopy = output.copy()
    
    TrainingData = data[:int(0.6 * len(dataCopy))]
    TrainingOutput = output[:int(0.6 * len(dataCopy))]

    RemainingData = data[int(0.6 * len(dataCopy)):]
    RemainingOutput = output[int(0.6 * len(dataCopy)):]

    ValidationData = RemainingData[:int(0.5*len(RemainingData))]
    ValidationOutput = RemainingOutput[:int(0.5*len(RemainingOutput))]

    TestingData = RemainingData[int(0.5 * len(RemainingData)):]
    TestingOutput = RemainingOutput[int(0.5 * len(RemainingData)):]

    return (TrainingData, TrainingOutput, ValidationData, ValidationOutput, 
    TestingData, TestingOutput)
    
```

FOR LSTM Use previous data strictly to predict time t. 
Example: t-5,t-4,t-3,t-2,t-1 to predict the t time. 
This makes sense logically because at time t, you will obtain the values on that day but we want to predict ahead of time! 

```python
def return_sequences(data, outputData, n_steps):
    X = list()
    Y = list() 
    length = len(data)
    for i in range(length):
        end_indx = i + n_steps 
        if (end_indx >= len(data)):
            break
        else:
            Xsample = data.iloc[i:end_indx, :]
            Ysample = outputData[end_indx-1]
            X.append(Xsample)
            Y.append(Ysample) 
    return (X, Y)

sequences = return_sequences(LSTMData, DemandData,3)
Xsequence = sequences[0]
Ysequence = sequences[1]

```
